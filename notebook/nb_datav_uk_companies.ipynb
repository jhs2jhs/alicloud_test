```python
import pandas
import numpy
```

```python
df_postcodes = pandas.read_csv('./data/postcodes.csv')
df_postcodes = df_postcodes[['Postcode', 'Latitude', 'Longitude', 'County']]
df_postcodes = df_postcodes.rename(columns={'Postcode': 'postcode', 'Latitude': 'latitude', 'Longitude': 'longitude', 'County': 'county'})
df_postcodes[df_postcodes['county'].notnull()]
```

```python
df_companies = pandas.read_csv('./data/BasicCompanyDataAsOneFile-2018-02-01.csv')
df_companies
```

```python
df_companies.columns
```

```python
df_companies[[' CompanyNumber', 'CompanyStatus']].groupby(by=['CompanyStatus']).agg(['count'])
```

```python
print('all available companies count: ', df_companies.shape[0])
df_companies = df_companies[df_companies['CompanyStatus'] == 'Active']
df_companies = df_companies[df_companies['RegAddress.PostCode'].notnull()]
print('active and registered only companies count: ', df_companies.shape)
```

```python
df_companies_simple = df_companies[['CompanyName', ' CompanyNumber', 'RegAddress.PostCode', 'CompanyCategory', 'IncorporationDate', 'SICCode.SicText_1', 'Accounts.AccountRefMonth']]
df_companies_simple
```

```python
df = pandas.merge(df_companies_simple, df_postcodes[['postcode', 'latitude', 'longitude', 'county']], how="left", left_on='RegAddress.PostCode', right_on="postcode")
df = df[(df['latitude'].notnull()) & (df['longitude'].notnull())]
df
```

```python
df.columns.tolist()
```

```python
rns = {
    'CompanyName': 'company_name',
    ' CompanyNumber': 'company_regis',
    'CompanyCategory': 'category',
    'IncorporationDate': 'incorp_date',
    'SICCode.SicText_1': 'sic_code',
    'Accounts.AccountRefMonth': 'financial_month',
}

df = df.drop('RegAddress.PostCode', axis=1).rename(columns=rns)
df
```

```python
df['sic_code'] = df['sic_code'].str.split('-').str.get(0).str.strip()
df
```

```python
df_sections = pandas.read_csv('./data/uk-sic-2007-condensed_csv.csv')
df_sections['sic_code'] = df_sections['sic_code'].astype(str)
df_sections
```

```python
df_sections.dtypes
```

```python
df = pandas.merge(df, df_sections[['sic_code', 'section', 'section_description']], on='sic_code', how='left')
df
```

```python
df = df.dropna()
print('clean dataset: ', df.shape)
```

```python
TABLES = {}
TABLES['uk_company'] = (
    "CREATE TABLE `uk_company` ("
    "  `company_id` varchar(10) NOT NULL,"
    "  `lat` DOUBLE NOT NULL,"
    "  `lng` DOUBLE NOT NULL,"
    "  `newness` varchar(16) NOT NULL,"
    "  `eastablished_year` int NOT NULL,"
    "  `section` varchar(1) NOT NULL,"
    "  PRIMARY KEY (`company_id`)"
    ") ENGINE=InnoDB")
TABLES['uk_county'] = (
    "CREATE TABLE `uk_county` ("
    "  `county_id` varchar(50) NOT NULL,"
    "  `count` int NOT NULL,"
    "  `total` int NOT NULL,"
    "  `percentage` double NOT NULL,"
    "  PRIMARY KEY (`county_id`)"
    ") ENGINE=InnoDB")
TABLES['uk_newness'] = (
    "CREATE TABLE `uk_newness` ("
    "  `county_id` varchar(50) NOT NULL,"
    "  `newness_type` varchar(50) NOT NULL,"
    "  `count` int NOT NULL,"
    "  PRIMARY KEY (`county_id`, `newness_type`)"
    ") ENGINE=InnoDB")
TABLES['uk_years'] = (
    "CREATE TABLE `uk_years`  ("
    "  `year_id` int NOT NULL,"
    "  `count` int NOT NULL,"
    "  PRIMARY KEY (`year_id`)"
    ") ENGINE=InnoDB")
TABLES['uk_section'] = (
    "CREATE TABLE `uk_section`  ("
    "  `section_id` varchar(50) NOT NULL,"
    "  `section_description` varchar(100) NOT NULL,"
    "  `count` int NOT NULL,"
    "  PRIMARY KEY (`section_id`)"
    ") ENGINE=InnoDB")
```

```python
def create_tables(cnx):
    cursor = cnx.cursor()
    for name in TABLES:
        ddl = TABLES[name]
        try:
            sql = 'DROP TABLE IF EXISTS `{}`'.format(name)
            #print(sql)
            cursor.execute(sql)
            cnx.commit()
            print("Creating table {}: ".format(name), end='')
            cursor.execute(ddl)
            cnx.commit()
        except mysql.connector.Error as err:
            if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
                print("already exists.")
            else:
                print(err.msg)
        else:
            print("OK")

    cursor.close()
    
create_tables(cnx)
```
